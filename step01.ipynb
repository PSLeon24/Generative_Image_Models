## 1. Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

def normal(x, mu=0, sigma=1): # mean = 0, standard deviation = 1
  y = 1 / (np.sqrt(2 ** np.pi) * sigma) * np.exp(-(x - mu) ** 2 / (2 * sigma**2))
  return y

x = np.linspace(-5, 5, 100)
y = normal(x)

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Standard Normal Distribution')

plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = normal(x, -3, 1)

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Normal Distribution')

plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = normal(x, 0, 5)

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Normal Distribution')

plt.show()
```

```python
x = np.linspace(-10, 10, 1000)

y0 = normal(x, mu=-3)
y1 = normal(x, mu=0)
y2 = normal(x, mu=5)

plt.plot(x, y0, label='$\mu$=-3')
plt.plot(x, y1, label='$\mu$=0')
plt.plot(x, y2, label='$\mu$=5')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Normal Distribution')
plt.legend()

plt.show()
```

```python
x = np.linspace(-10, 10, 1000)

y0 = normal(x, sigma=0.5)
y1 = normal(x, sigma=1)
y2 = normal(x, sigma=2)

plt.plot(x, y0, label='$\sigma$=0.5')
plt.plot(x, y1, label='$\sigma$=1')
plt.plot(x, y2, label='$\sigma$=2')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Normal Distribution')
plt.legend()

plt.show()
```

## 2. Central Limit Theorem

```python
N = 3

xs = []
for n in range(N):
  x = np.random.rand()
  xs.append(x)

x_mean = np.mean(xs)
print(f'mean of x is {x_mean:.3f}')
```

```python
x_means = []
N = 1

for _ in range(10000):
  xs = []
  for n in range(N):
    x = np.random.rand()
    xs.append(x)
  x_mean = np.mean(xs)
  x_means.append(x_mean)

plt.hist(x_means, bins='auto', density=True)

plt.xlabel('x')
plt.ylabel('frequency')
plt.title('Probability Density')

plt.xlim(-0.05, 1.05)
plt.ylim(0, 5)

plt.show()
```

```python
x_sums = []
N = 5

for _ in range(10000):
  xs = []
  for n in range(N):
    x = np.random.rand()
    xs.append(x)
  x_sum = np.sum(xs)
  x_sums.append(x_sum)

x_norm = np.linspace(-5, 5, 1000)
mu = N / 2
sigma = np.sqrt(N / 12)
y_norm = normal(x_norm, mu, sigma)

plt.hist(x_sums, bins='auto', density=True)
plt.plot(x_norm, y_norm, color='red')

plt.xlabel('x')
plt.ylabel('frequency')
plt.title(f'N={N}')
plt.xlim(-1, 6)
plt.show()
```
